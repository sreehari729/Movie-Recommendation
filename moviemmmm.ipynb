```python
#converting the string to list
import ast
def convert(text):
    l =[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
# now removing space like that

def remove_space(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

```python
# Concatinating
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# droping extra columns
new_data = movies[['movie_id','title','tags']]
new_data.head()
```

```python
# Converting list to str
new_data['tags'] = new_data['tags'].apply(lambda x: " ".join(x))
new_data.head()
```

```python
new_data.iloc[0]['tags']
```

```python
# Converting to lower case
new_data['tags'] = new_data['tags'].apply(lambda x:x.lower())
```

```python
new_data.shape
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    T = []

    for i in text.split():
        T.append(ps.stem(i))

    return " ".join(T)
```

```python
new_data['tags'] = new_data['tags'].apply(stems)
```

```python
new_data.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_data['tags']).toarray()
```

```python
vector[0]
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
new_data[new_data['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new_data[new_data['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_data.iloc[i[0]].title)
```

```python
recommend('Spider-Man 2')
```

```python
recommend('The Dark Knight Rises')
```
